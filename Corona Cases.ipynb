```python
import pandas as pd #Uses Pandas Library
import seaborn as sns #Uses Seaborn Library

#Grabs excel data from url site and converts excel to a pandas readable matrix
url = 'https://www.ecdc.europa.eu/sites/default/files/documents/COVID-19-geographic-disbtribution-worldwide-2020-03-26.xlsx'
df_raw = pd.read_excel(url)

#Returns a list of Countries to choose from
df_raw['Countries and territories'].unique()
```

```python
#Declare new instance copy of df_raw
df1 = df_raw

#Reinstatiates instance to only contain data with 'Countries and Territories' named 'United_States_of_America'
df1 = df1[df1['Countries and territories']=='United_States_of_America']

#Sorts the reinstantiated data based on Year, Month, Day
df1 = df1.sort_values(['Year', 'Month' ,'Day'], ascending=[1, 1, 1])

#Sets datetime dataframe from Pandas into the dataset
df1['ts'] = pd.to_datetime(df1[['Year', 'Month' ,'Day']])


################ My Take - Filtered using Philippines
#Declare new instance copy of df_raw
df2 = df_raw

#Reinstatiates instance to only contain data with 'Countries and Territories' named 'Philippines'
df2 = df2[df2['Countries and territories']=='Philippines']

#Sorts the reinstantiated data based on Year, Month, Day
df2 = df2.sort_values(['Year', 'Month' ,'Day'], ascending=[1, 1, 1])

#Sets datetime dataframe from Pandas into the dataset
df2['ts'] = pd.to_datetime(df2[['Year', 'Month' ,'Day']])


```

```python
#Create Plot based on variable df (which is the dataframe that was declared and repurposed)
# And puts into and readable plot chart format
chart = sns.lineplot(x='ts', y='Cases', data=df1, hue='Countries and territories')
chart.set_xticklabels(df1['ts'],rotation=45)
```

```python
#This Chart shows the data for the Philippines
chart2 = sns.lineplot(x='ts', y='Cases', data=df2, hue='Countries and territories')
chart2.set_xticklabels(df1['ts'],rotation=45)
```

```python
#This Data shows a comparison between the United States and The Philippines
chart = sns.lineplot(x='ts', y='Cases', data=df1, hue='Countries and territories')
chart.set_xticklabels(df1['ts'],rotation=45)
chart2 = sns.lineplot(x='ts', y='Cases', data=df2, hue='Countries and territories')
chart2.set_xticklabels(df1['ts'],rotation=45)
```
